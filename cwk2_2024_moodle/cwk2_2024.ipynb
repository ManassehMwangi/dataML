```python
## Code for Part 1
import pandas as pd
import numpy as np
# Load the training dataset
df = pd.read_json('UKLoad2023.json')

###################################################
## Write your code for Task 1.2 here. 
###################################################
## Write code to validate the integrity of the dataset by checking for duplicate data and applying appropriate solutions.

#
# Check for duplicate entries
duplicates = df.duplicated()
print(f"Number of duplicate entries: {duplicates.sum()}")

# If there are duplicates, we can remove them
if duplicates.sum() > 0:
    df = df.drop_duplicates()
    print("Duplicates have been removed.")

# Check for duplicates based on specific columns
# We'll consider entries as duplicates if they have the same settlementDate, settlementPeriod, and quantity
specific_duplicates = df.duplicated(subset=['settlementDate', 'settlementPeriod', 'quantity'])
print(f"Number of specific duplicates: {specific_duplicates.sum()}")

# If there are specific duplicates, we can remove them
if specific_duplicates.sum() > 0:
    df = df.drop_duplicates(subset=['settlementDate', 'settlementPeriod', 'quantity'])
    print("Specific duplicates have been removed.")

# Verify the final shape of the dataset
print(f"Final shape of the dataset: {df.shape}")

# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)

# If there are missing values, we might want to handle them
if missing_values.sum() > 0:
    # For this dataset, we'll drop rows with missing values
    # You might want to use a different strategy depending on your specific needs
    df = df.dropna()
    print("Rows with missing values have been removed.")

print(f"\nFinal shape of the dataset after handling missing values: {df.shape}")

###################################################
## Write your code for Task 1.3 here. 
###################################################
## Write code to validate the integrity of the dataset by checking missing values (NaNs) and applying appropriate solutions.

# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)

# Calculate the percentage of missing values
missing_percentage = (missing_values / len(df)) * 100
print("\nPercentage of missing values in each column:")
print(missing_percentage)

# Identify columns with missing values
columns_with_missing = missing_values[missing_values > 0].index.tolist()

if len(columns_with_missing) > 0:
    print(f"\nColumns with missing values: {columns_with_missing}")
    
    for column in columns_with_missing:
        print(f"\nHandling missing values in column: {column}")
        
        # Check the data type of the column
        dtype = df[column].dtype
        
        if dtype in ['int64', 'float64']:
            # For numeric columns, we'll fill with the median
            median_value = df[column].median()
            df[column].fillna(median_value, inplace=True)
            print(f"Filled missing values with median: {median_value}")
        elif dtype == 'object':
            # For categorical columns, we'll fill with the mode
            mode_value = df[column].mode()[0]
            df[column].fillna(mode_value, inplace=True)
            print(f"Filled missing values with mode: {mode_value}")
        else:
            # For other types (like datetime), we'll drop the rows
            df.dropna(subset=[column], inplace=True)
            print(f"Dropped rows with missing values in this column")

    # Verify that all missing values have been handled
    final_missing = df.isnull().sum().sum()
    print(f"\nRemaining missing values after handling: {final_missing}")
else:
    print("\nNo missing values found in the dataset.")

print(f"\nFinal shape of the dataset: {df.shape}")
###################################################
## Write your code for Task 1.5 here. 
###################################################
# Write code to prepare the following two dataset: the feature dataset with the name of "feature_dataset", which should be 
# a 2-dimensional numpy ndarray format, each row is a different day, and each row contains 48 values from the field "quantity" 
# in that day; the label dataset with the name of "label_dataset", which is with a similar format of the "feature_dataset", 
# but the date for each row should be one day later. For example, if row 3 in "feature_dataset" is the data for 2023-8-20, 
# then row 3 in "label_dataset" should be the data for 2023-8-21. This will prepare the "feature_dataset" and "label_dataset"
# ready for the day-ahead model training and validation purpose.

# dataset initialization - you might need to rewrite/remove this part if necessary
# feature_dataset = np.ones((10,10)) # - you might need to rewrite/remove this part if necessary
# label_dataset   = np.ones((10,10)) # - you might need to rewrite/remove this part if necessary

#
# Sort the dataframe by date and settlement period
df['startTime'] = pd.to_datetime(df['startTime'])
df = df.sort_values(['startTime', 'settlementPeriod'])

# Group the data by date
daily_data = df.groupby(df['startTime'].dt.date)

# Initialize lists to store feature and label data
feature_list = []
label_list = []

# Iterate through the grouped data
for date, group in daily_data:
    # Check if we have exactly 48 settlement periods for this day
    if len(group) == 48:
        # Add this day's data to the feature list
        feature_list.append(group['quantity'].values)
        
        # Get the next day's data for the label
        next_day = date + pd.Timedelta(days=1)
        if next_day in daily_data.groups:
            next_day_data = daily_data.get_group(next_day)
            if len(next_day_data) == 48:
                label_list.append(next_day_data['quantity'].values)
            else:
                # If next day doesn't have 48 periods, remove this day from features
                feature_list.pop()

# Convert lists to numpy arrays
feature_dataset = np.array(feature_list)
label_dataset = np.array(label_list)

# Ensure feature_dataset and label_dataset have the same number of rows
min_rows = min(feature_dataset.shape[0], label_dataset.shape[0])
feature_dataset = feature_dataset[:min_rows]
label_dataset = label_dataset[:min_rows]

print("Shape of feature_dataset:", feature_dataset.shape)
print("Shape of label_dataset:", label_dataset.shape)

# Display the first few rows of each dataset
print("\nFirst few rows of feature_dataset:")
print(feature_dataset[:3])
print("\nFirst few rows of label_dataset:")
print(label_dataset[:3])

# Verify the date shift
print("\nVerifying date shift:")
print("First day in feature_dataset:", df['startTime'].dt.date.min())
print("First day in label_dataset:", df['startTime'].dt.date.min() + pd.Timedelta(days=1))
###################################################
## Write your code for Task 1.6 here. 
###################################################
# Write code to visualize the day-ahead forecasting problem, by plotting the day-ahead data from  "feature_dataset" and the corresponding output data from "label_dataset".

#

import matplotlib.pyplot as plt

# Choose a random day to visualize (you can change this index)
day_index = np.random.randint(0, feature_dataset.shape[0])

# Create a time array for x-axis (48 half-hour periods)
time = np.arange(0, 24, 0.5)

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(time, feature_dataset[day_index], label='Current Day', color='blue')
plt.plot(time, label_dataset[day_index], label='Next Day (Target)', color='red', linestyle='--')

plt.title(f'Day-Ahead Load Forecasting - Sample Day {day_index}')
plt.xlabel('Hour of the Day')
plt.ylabel('Load (MW)')
plt.legend()
plt.grid(True)

# Add markers for each half-hour
plt.xticks(np.arange(0, 24.5, 2))  # Major ticks every 2 hours
plt.gca().set_xticks(np.arange(0, 24.5, 1), minor=True)  # Minor ticks every hour

plt.tight_layout()
plt.show()

# Calculate and print some statistics
print(f"Sample day index: {day_index}")
print(f"Current day mean load: {feature_dataset[day_index].mean():.2f} MW")
print(f"Next day mean load: {label_dataset[day_index].mean():.2f} MW")
print(f"Current day peak load: {feature_dataset[day_index].max():.2f} MW")
print(f"Next day peak load: {label_dataset[day_index].max():.2f} MW")

# Calculate correlation between current day and next day
correlation = np.corrcoef(feature_dataset[day_index], label_dataset[day_index])[0, 1]
print(f"Correlation between current day and next day: {correlation:.4f}")
```
